```python
p = float(input("Введите число p: "))
q = float(input("Введите число q: "))


a = p * (q**2)
b = (p**2) + (p * q)


if a > b:
    c = a * (a - b)
elif a == b:
    c = a**3
else:
    c = p - q


d = c * (p + q)

print(f"Результат d = {d}")
```

```python
from math import *
from matplotlib.pyplot import *

a = float(input())
b = float(input())
h = float(input())

xs = []
ys = []
rows = []

n = int((b - a) / h) + 1

for i in range(n):
    x = round(a + i * h, 10)
    if x > b + 1e-9: break
    
    c = cos(x)
    if abs(c) > 1e-10:
        f = x / c
        xs.append(x)
        ys.append(f)
        rows.append([x, round(f, 4)])

fig, (ax1, ax2) = subplots(1, 2, figsize=(10, 5))

ax1.plot(xs, ys, marker='.')
ax1.set_title("F(x) = x / cos(x)")
ax1.grid()

ax2.axis('off')
table(cellText=rows, colLabels=['x', 'F(x)'], loc='center')

show()
```
