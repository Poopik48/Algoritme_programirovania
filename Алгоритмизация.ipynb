```python
class LampRow:
    def __init__(self, count):
        self.__count = count
        self.__state = "0" * count

    @property
    def state(self):
        return self.__state

    @state.setter
    def state(self, value):
        if isinstance(value, str) and len(value) == self.__count:
            self.__state = value
        else:
            self.__state = "0" * self.__count

    def show(self):
        print("".join("*" if ch == "1" else "-" for ch in self.__state))


lamps = LampRow(6)
lamps.show()
lamps.state = "101010"
print(lamps.state)
lamps.show()
lamps.state = "10101010"
print(lamps.state)
lamps.show()
```

```python
class LampRow:
    def __init__(self, count):
        self.__count = count
        self.__state = "0" * count

    @property
    def state(self):
        return self.__state

    @state.setter
    def state(self, value):
        if (
            isinstance(value, str)
            and len(value) == self.__count
            and all(ch in "012" for ch in value)
        ):
            self.__state = value
        else:
            self.__state = "0" * self.__count

    def show(self):
        result = ""
        for ch in self.__state:
            if ch == "1":
                result += "*"
            elif ch == "2":
                result += "o"
            else:
                result += "-"
        print(result)


lamps = LampRow(6)
lamps.show()
lamps.state = "102102"
print(lamps.state)
lamps.show()
lamps.state = "10201010"
print(lamps.state)
lamps.show()
```

```python
class LampRow:
    def __init__(self, count):
        self.__count = count
        self.__state = 0

    @property
    def state(self):
        return str(self.__state).zfill(self.__count)

    @state.setter
    def state(self, value):
        if (
            isinstance(value, str)
            and len(value) == self.__count
            and all(ch in "012" for ch in value)
        ):
            self.__state = int(value)
        else:
            self.__state = 0

    def show(self):
        s = self.state
        result = ""
        for ch in s:
            if ch == "1":
                result += "*"
            elif ch == "2":
                result += "o"
            else:
                result += "-"
        print(result)


lamps = LampRow(6)
lamps.show()
lamps.state = "102102"
print(lamps.state)
lamps.show()
lamps.state = "10201010"
print(lamps.state)
lamps.show()
```
